```python
# Import Dependencies "Resources/threatened_species.csv"
import pandas as pd
from pathlib import Path
```

```python
# Display the DataFrame
organised_world_data_path = Path("combine/organised_world_data.csv")
organised_world_data_df = pd.read_csv(organised_world_data_path)
organised_world_data_df.head()
```

```python
# Display the DataFrame
country_air_path = Path("combine/country_air_quality.csv")
country_air_df = pd.read_csv(country_air_path)
country_air_df.head()
```

```python
# Display the DataFrame
death_air_path = Path("combine/death_air_pollution.csv")
death_air_df = pd.read_csv(death_air_path)
death_air_df.head()
```

```python
# Merge the three DataFrames together based on the Country they share
world_df = pd.merge(organised_world_data_df, country_air_df, on=["Country"], how="inner")
world_df = world_df.merge(death_air_df, on=["Country"], how="inner")
world_df.head(167)
```

```python
world_df.columns
```

```python
world_df.to_csv("merged_world_data.csv",index=False,header=True)
```
