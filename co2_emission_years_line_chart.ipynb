```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Display the DataFrame
co2_emission_years_path = Path("co2_emission_years.csv")
co2_emission_years_df = pd.read_csv(co2_emission_years_path)
co2_emission_years_df.head()
```

```python
#df.plot.line(x=CO‚ÇÇ Emission(2000), CO‚ÇÇ Emission(2005), CO‚ÇÇ Emission(2010), CO‚ÇÇ Emission(2015), CO‚ÇÇ Emission(2020), CO‚ÇÇ Emission(2021))
```

```python
co2_emission_years_df=co2_emission_years_df.loc[co2_emission_years_df['Country']!='World']
```

```python
co2_emission_years_df=co2_emission_years_df.set_index('Country')
co2_emission_years_df.head()
```

```python
co2_emission_years_df['average emission']=co2_emission_years_df.mean(axis=1)
co2_emission_years_df.head()
```

```python
Top_emission_countires_df=co2_emission_years_df.sort_values('average emission',ascending=False).head()
del Top_emission_countires_df['average emission']
Top_emission_countires_df
```

```python
Bottom_emission_countires_df=co2_emission_years_df.sort_values('average emission').head()
del Bottom_emission_countires_df['average emission']
Bottom_emission_countires_df
```

```python
Top_emissions_countires_plot_df=pd.DataFrame({
    'year':[2000,2005,2010,2015,2020,2021],
    'China':list(Top_emission_countires_df.iloc[0]),
    'United States':list(Top_emission_countires_df.iloc[1]),
    'India':list(Top_emission_countires_df.iloc[2]),
    'Russia':list(Top_emission_countires_df.iloc[3]),
    'Japan':list(Top_emission_countires_df.iloc[4])
    
    
})
#Top_emissions_countires_plot_df['year']=pd.to_datetime(Top_emissions_countires_plot_df['year'])
Top_emissions_countires_plot_df.set_index('year',inplace=True)
Top_emissions_countires_plot_df.plot(style='.-')
plt.ylabel('CO₂ Emission (tonnes)')
plt.title('CO₂ Emission (tonnes) for Countries with Highest Emission')
plt.show()
```

```python
Bottom_emissions_countires_plot_df=pd.DataFrame({
    'year':[2000,2005,2010,2015,2020,2021],
    'Niue':list(Bottom_emission_countires_df.iloc[0]),
    'Tuvalu':list(Bottom_emission_countires_df.iloc[1]),
    'Saint Helena':list(Bottom_emission_countires_df.iloc[2]),
    'Wallis and Futana':list(Bottom_emission_countires_df.iloc[3]),
    'Montserrat':list(Bottom_emission_countires_df.iloc[4])
    
    
})
#Top_emissions_countires_plot_df['year']=pd.to_datetime(Top_emissions_countires_plot_df['year'])
Bottom_emissions_countires_plot_df.set_index('year',inplace=True)
Bottom_emissions_countires_plot_df.plot(style='.-')
plt.ylabel('CO₂ Emission (tonnes)')
plt.title('CO₂ Emission (tonnes) for Countries with Lowest Emission')
plt.show()
```
